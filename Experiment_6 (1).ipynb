```python
''' 6) Write a Python program to calculate the gross salary of an employee based
If basic <= 10000: HRA = 20% , DA = 80%
If basic <=20000: HRA = 25% , DA = 90%
Otherwise : HRA = 30% , DA = 95%'''
basic = int(input("Enter salary: "))
if basic <=10000:
  hra = basic*0.20
  da = basic*0.80
elif basic<=20000:
  hra = basic*0.25
  da = basic*0.90
else:
  hra = basic*0.30
  da = basic *0.95

gross = basic + hra + da
print("Gross Salary:", gross)
```

```python
''' 7) Write a Python program to calculate income tax based on annual income salary
Upto 250000 : No Tax
250001 - 500000 : 5%
500001 - 1000000 :20%
Above 1000000 : 30%'''
sal = int (input("Enter salary:"))
if sal<=250000:
  tax = 0
elif sal<=500000:
  tax = (sal - 250000)*0.05
elif sal<=1000000:
  tax = (sal-500000)*0.20 + 250000*0.05
else:
  tax = (sal-1000000)*0.30 + 500000*0.20 + 250000*0.05

print("Total Income Tax:",tax)
```
